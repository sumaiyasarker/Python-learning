Q1. Simple Calculator Program

Write a Python program that performs basic arithmetic operations.

```python
# --- Part A: Input & Setup ---

# 1. Ask the user to input two numbers
print("--- Input ---")
num1_input = input("Enter the first number: ")
num2_input = input("Enter the second number: ")
```

```python
# 2. Convert both inputs into floats (using float allows for decimal numbers)
number1 = float(num1_input)
number2 = float(num2_input)
```

```python
# --- Part B: Calculations ---

# Compute the following operations:

# 1. Addition
addition = number1 + number2

# 2. Subtraction
subtraction = number1 - number2

# 3. Multiplication
multiplication = number1 * number2

# 4. Division
# I add a simple check to prevent the program from crashing if the user divides by zero
if number2 != 0:
    division = number1 / number2
else:
    division = "Undefined (Cannot divide by zero)"

# 5. Clearly labeled outputs
print("\n--- Calculation Results ---")
print(f"Addition:       {number1} + {number2} = {addition}")
print(f"Subtraction:    {number1} - {number2} = {subtraction}")
print(f"Multiplication: {number1} * {number2} = {multiplication}")
print(f"Division:       {number1} / {number2} = {division}")
# --- Part C: Code Style ---
# (Comments have been used throughout to explain steps)
# (Proper spacing and indentation have been applied)
```

Q3. Temperature Converter.
Write a program to convert Celsius to Fahrenheit.

```python
# --- Part A: Input ---
# 1. Take temperature in Celsius from the user.
celsius_input = input("Enter temperature in Celsius: ")

# 2. Convert to float.
# I convert to float so we can perform decimal calculations.
celsius_float = float(celsius_input)
```

```python
# --- Part B: Calculation ---
# Formula: F = (C * 9/5) + 32
#Use the variable 'celsius_float' created in Part A.
fahrenheit = (celsius_float * 9/5) + 32
```

```python
# --- Part C: Output ---
# Print Fahrenheit with a clear message.
#Use an f-string for clean formatting.
# The ° symbol is included to match the example in the assignment.
print(f"Temperature in Fahrenheit is: {fahrenheit}°F")
```

Q4. Billing Program.
Write a program that calculates a shopping bill.

```python
# --- Part A: Inputs ---
# 1. Ask for Item Name
item_name = input("Enter the item name: ")

# 2. Ask for Price
# Use float() because prices can have decimal points (e.g., 10.50).
price = float(input("Enter the price of the item: "))

# 3. Ask for Quantity
# Use int() because quantity is usually a whole number.
quantity = int(input("Enter the quantity: "))
```

```python
# --- Part B: Calculation---
# Formula: Total = price * quantity
total_bill = price * quantity
```

```python
# --- Part C: Output ---
# The format requested is: Total Bill for [Item Name] = [Total]
# Use an f-string for clear formatting.
print(f"Total Bill for {item_name} = {total_bill}")
```
